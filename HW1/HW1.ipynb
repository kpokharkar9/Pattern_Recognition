## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


Matrix T is essentially a rearranged version of the identity matrix, where the order of its rows has been shuffled according to the rows array. 

In other words, T is a permutation of the rows of the identity matrix, maintaining the same elements but changing their order based on the random row indices.

```python
# Create matrix T by rearranging the rows of identity_matrix using the shuffled row indices
T_rows = identity_matrix[rows]

# Check the shape of T_rows and reshape it if needed to match the size of img_average_channel
if T_rows.shape != img_average_channel.shape:
    T_rows = T_rows.reshape(img_average_channel.shape)

# Print the shape of T
print("Shape of T_rows:", T_rows.shape)
```

```python
# Perform matrix multiplication between T_rows and img_average_channel
transformed_image_rows = np.dot(T_rows, img_average_channel)

# Check the shape of the result
print("Shape of transformed_image:", transformed_image_rows.shape)

# Plot transformed_image
plt.imshow(transformed_image_rows)
plt.title("Transformed Image")
plt.show()
```

The dots in the transformed image correspond to individual pixel values. Each pixel retains its value, but its position within the rows may have changed.

2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
# Check if T is square and has full rank
if T_rows.shape[0] == T_rows.shape[1] and np.linalg.matrix_rank(T_rows) == T_rows.shape[0]:
    print("Matrix T is invertible.")
else:
    print("Matrix T is not invertible.")
    
# Check if T is square and has full rank
if T_rows.shape[0] == T_rows.shape[1] and np.linalg.matrix_rank(T_rows) == T_rows.shape[0]:
    # Compute the inverse of T
    Tinv_rows = np.linalg.inv(T_rows)

    # Perform the inverse operation to retrieve the original image
    retrieved_image_rows = np.dot(Tinv_rows, transformed_image_rows)

    # Plot the retrieved original image
    plt.imshow(retrieved_image_rows)
    plt.title("Retrieved Original Image")
    plt.show()
else:
    print("Matrix T is not invertible, and the inverse cannot be computed.")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
image_shape = image.shape[:2]

# Create identity matrices with the same shape as the image
identity_matrix_rows = np.eye(*image_shape)
identity_matrix_cols = np.eye(*image_shape)

# Shuffle the row and column indices
random_row_indices = np.arange(image_shape[0])
np.random.shuffle(random_row_indices)

random_col_indices = np.arange(image_shape[1])
np.random.shuffle(random_col_indices)

# Create shuffled matrices T_rows and T_cols
T_rows = identity_matrix_rows[random_row_indices]
T_cols = identity_matrix_cols[:, random_col_indices]

# Initialize an empty array for the scrambled image
scrambled_image = np.empty_like(image)

# Apply the scrambling operations to each color channel separately
for channel in range(image.shape[2]):
    scrambled_image[:, :, channel] = np.dot(np.dot(T_rows, image[:, :, channel]), T_cols)

# Plot the scrambled image
plt.imshow(scrambled_image)
plt.title("Scrambled Image")
plt.show()

# Check if T_rows and T_cols are invertible
if np.linalg.matrix_rank(T_rows) == T_rows.shape[0] and np.linalg.matrix_rank(T_cols) == T_cols.shape[0]:
    print("Matrices T_rows and T_cols are invertible.")
else:
    print("Matrices T_rows and T_cols are not invertible.")

# Invert the scrambling operations for each color channel separately
T_rows_inv = np.linalg.inv(T_rows)
T_cols_inv = np.linalg.inv(T_cols)

# Initialize an empty array for the recovered image
recovered_image = np.empty_like(image)

# Recover the original image by applying the inverse operations to each color channel
for channel in range(image.shape[2]):
    recovered_image[:, :, channel] = np.dot(np.dot(T_rows_inv, scrambled_image[:, :, channel]), T_cols_inv)

# Plot the recovered original image
plt.imshow(recovered_image)
plt.title("Recovered Original Image")
plt.show()
```
